```python
Data = np.array(df)
Data = list(Data)
for i in Data:
    i = list(i)
Data
```

```python
def prior_prob(Data):
    tot = len(Data)
    y,n=0.0,0.0
    for i in Data:
        if i[11] == 'Y':
            y+=1
        else:
            n+=1
    cardinality = []
    for i in df:
        cardinality.append(list(df[i].unique()))
    return y/tot,n/tot,y,n,cardinality
```

```python
PY,PN,CY,CN,cardinality = prior_prob(Data)
print(f'{PY}  {PN}  {CY}  {CN}')
```

```python
#likelihood
def likelihood(Data,PY,PN,CY,CN,cardinality):
    N_table = {}
    Y_table = {}
    for i in range(11):
        likeliy = {}
        likelin = {}
        if len(cardinality[i])<5:
            for j in cardinality[i]:
                n = 0.0
                y = 0.0
                for k in Data:
                    if k[i] == j and k[11] == 'N':
                        n+=1
                    elif k[i] == j and k[11] == 'Y':
                        y+=1
                y/=CY
                n/=CN
                print(f'{j}\t\t -- > \t{y:.4f} \t {n:.4f}')
                likeliy[j] = y if y!=0 else 0.003
                likelin[j] = n if n!=0 else 0.003
            N_table[i] = likelin
            Y_table[i] = likeliy
        else:
            ncol = []
            ycol = []
            for j in Data:
                if j[11] == 'N':
                    ncol.append(j[i])
                elif j[11] == 'Y':
                    ycol.append(j[i])
                
            nmean = np.mean(ncol)
            nvariance = sum([((x - nmean) ** 2) for x in ncol]) / len(ncol)
            nstd = nvariance ** 0.5
            ymean = np.mean(ycol)
            yvariance = sum([((x - ymean) ** 2) for x in ycol]) / len(ycol)
            ystd = yvariance ** 0.5
            likeliy['ystd'] = ystd
            likeliy['ymean'] = ymean
            likelin['nstd'] = nstd
            likelin['nmean'] = nmean
            N_table[i] = likelin
            Y_table[i] = likeliy
            
    return Y_table,N_table
Y_table,N_table = likelihood(Data,PY,PN,CY,CN,cardinality)
```

```python
def Posterior(query,Y_table,N_table):
    y = 1
    n = 1
    for i in range(len(query)-1):
        if len(cardinality[i])<5:
            y*=Y_table[i][query[i]]
            n*=N_table[i][query[i]]
        else:
            nstd = N_table[i]['nstd']
            nmean = N_table[i]['nmean']
            ystd = Y_table[i]['ystd']
            ymean = Y_table[i]['ymean']
            yx = query[i]
            nx = query[i]
            yprob = (1/(ystd*np.sqrt(2*np.pi)))*np.exp((-1*((yx-ymean)**2))/(2*(ystd**2)))
            nprob = (1/(nstd*np.sqrt(2*np.pi)))*np.exp((-1*((nx-nmean)**2))/(2*(nstd**2)))
            y*=yprob
            n*=nprob            
    y*=PY
    n*=PN
    
#     print(f'P(YES) : {y}   P(NO) : {n}')
#     if y>n:
#         print('Loan Status : YES')
#     else:
#         print ('Loan Status : No')
    
    return y,n
```

```python
def NB(Data):
    Result = []
    PY,PN,CY,CN,cardinality = prior_prob(Data)
    Y_table,N_table = likelihood(Data,PY,PN,CY,CN,cardinality)
    for query in Data:
        y,n = Posterior(query,Y_table,N_table)
        if y>n:
            Result.append('Y')
        else:
            Result.append('N')
    return Result
    
    
```

```python
X = np.array(df.iloc[:,:])
Y = np.array(df.iloc[:,11])
```

```python
ypred = NB(X)
ypred
```
